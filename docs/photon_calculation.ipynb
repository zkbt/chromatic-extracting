# Photon Calculation
This tutorial will use bias and flat exposures to find the photons per pixel. The following equation is used to solve for the photons, S:

P_i,j = SgF + B

where P_i,j is the pixels of the original image, S is the photons, g is the gain, F is the flat image pixels, and B is the bias image pixels.
