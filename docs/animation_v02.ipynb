In order for the animation to show a transit correctly, the fits files need to be organized in order of their observation. To do so, run:

```python
sorted_list = sorted_datetime()
```

This will create a list containing sorted information for the file name, observation date-time, header data, and image data.

If you created the animation now, you would see that while all of the fits files are in order, there are images that stand out and do not appear to belong in the overall set. This occurs due to the 'OBJECT' header. To ensure images flow smoothly, use the filtered_sorted_list() function and declare the object that you wish to look at:

```python
WASP_filter = filtered_sorted_list(sorted_list, 'WASP-94')
```

Now, the sorted_list has been further filtered and only includes information for the object chosen, in this case, WASP-94.

The animation can now be displayed using the animation() function. This function takes in many parameters so refer to the documentation in main_zasha.py. In this example, the defaults remain unchanged and the diagnostic information shown includes everything in red:

```python
# Set the display backend to support interactive figures
%matplotlib notebook
sorted_list_animation = WASP_filter
animation(sorted_list_animation, 'filename', 'object', 'exptime', 'airmass', 'ra', 'dec')
```
